```python
## Lab 3 by Andres Candido

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the seed for testing
np.random.seed(12)

X = np.random.normal(loc=10, scale=5, size=50000)

Y = np.random.gamma(shape=1.4, scale=20, size=50000)

# Set up the figure and gridspec
fig = plt.figure(figsize=(8, 8)) # create a figure of 8 x 8 inches (This is only used for displaying the fig, does not affect the fig itself)
gs = fig.add_gridspec(3, 3)      # defines grid of cells in a subplot


# Define the axes for the contour plot and marginal histograms
ax_joint = fig.add_subplot(gs[1:, :-1])
ax_marg_x = fig.add_subplot(gs[0, :-1], sharex=ax_joint)
ax_marg_y = fig.add_subplot(gs[1:, -1], sharey=ax_joint)

# Create the contour plot with 2D histogram
sns.histplot(x=X, y=Y, bins=50, pthresh=0.1, cmap="magma", ax=ax_joint)

# Create the marginal histograms
sns.histplot(x=X, bins=50, color="b", ax=ax_marg_x)
sns.histplot(y=Y, bins=50, color="r", ax=ax_marg_y)

# Set the axis labels and title
ax_joint.set_xlabel("X")
ax_joint.set_ylabel("Y")
ax_marg_x.set_ylabel("Frequency")
ax_marg_y.set_xlabel("Frequency")
fig.suptitle("Interaction between X and Y")

# Show the plot
plt.show()
```
