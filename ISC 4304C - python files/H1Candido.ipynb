4. Sum up 10 random number (look at our calculation of Pi for that) and present them in a list, use a ’for’ or ’while’ loop!

```python
import random

nums = []
num_sum = 0

for i in range(10):
  temp = random.randint(0, 100)
  num_sum = num_sum + temp
  nums.append(temp)

print("List of random numbers:\n",nums)
print("Sum of those numbers: ",num_sum)
```

5. Present the code and calculation for this math formula:

However, we cannot sum up an infinite number of values, show results where you sum over
K = 10, K = 1000, K = 10000

```python
def sum_formula(K):
  sum = 0
  All_ks = list(range(1,K+1))
  for i in range(K):
    sum = sum + (1/All_ks[i])   
  return(sum)

# when K = 10
print(sum_formula(10))

# when K = 1000
print(sum_formula(1000))

# when K = 10000
print(sum_formula(10000))
```
